```python
import nltk
```

```python
# use to convert the words love,loving,loves to love (stemming)
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
ps.stem('doing')
```

```python
movi['tags']=movi['tags'].apply(stem)
```

```python
movi['tags'].values
```

```python
# check the distance between the vectors by cosine_similarity but not eclide distance(not reccomended)
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
movi[movi['Movie']=='Arjun Reddy'].index[0]
```

```python
sorted(similarity[0],reverse=True)
```

```python
# function recommd
def recommend(movies):
    movie_index=movi[movi['Movie']==movies].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(movi.iloc[i[0]].Movie)
```

```python
movi['Movie'].head(10)
```

```python
recommend('Arjun Reddy')
```

```python
recommend('Rangasthalam')
```

```python
recommend('Arjun Reddy')
```
